```python
import random
import copy
Q_table = {}
alpha = 0.1
gamma = 1
epsilon = 0.5



def available_action(state):
  lis = []

  for i,row in enumerate(state):
    for j,val in enumerate(row):
      if val == 0:
        x = i
        y = j
  if x>0:
    lis.append(1)
  if x<2:
    lis.append(2)
  if y>0:
    lis.append(3)
  if y<2:
    lis.append(4)
  return lis

def apply_action(state2, a):

  # state = state2.copy()  #ye gare just copy outer list. so use deep copy
  state = copy.deepcopy(state2)
  for i,row in enumerate(state):
    for j,val in enumerate(row):
      if val == 0:
        x = i
        y = j


  if a == 1:
      state[x][y], state[x-1][y] = state[x-1][y], state[x][y]
  if a == 2:
      state[x][y], state[x+1][y] = state[x+1][y], state[x][y]
  if a == 3:
      state[x][y], state[x][y-1] = state[x][y-1], state[x][y]
  if a == 4:
      state[x][y], state[x][y+1] = state[x][y+1], state[x][y]

  return state

def reward(s):
  if s == [[1,2,3],[4,5,6],[7,8,0]]:
    return 1000
  else:
    return -1

for episode in range(1000):
  state = [[1,8,2],[0,4,3],[7,6,5]]
  for i in range(1000):
    choose_action = available_action(state)
    if random.random() < epsilon:
      action = random.choice(choose_action)
    else:
      max_q = -1000000
      st = []
      for a in choose_action:
        q_value = Q_table.get((tuple(tuple(row) for row in state),a),0)
        max_q = max(max_q,q_value)
        st.append((q_value,a))
      best_actions = []
      for q,a in st:
        if(q == max_q):
          best_actions.append(a)

      action = random.choice(best_actions)

    next_state = apply_action(state,action)
    choose_action1 = available_action(next_state)
    r = reward(next_state)

    Q_max = 0
    for a in choose_action1:
      Q_max = max(Q_max,Q_table.get((tuple(tuple(row) for row in next_state), a), 0))
    Q_table[(tuple(tuple(row) for row in state), action)] = Q_table.get((tuple(tuple(row) for row in state), action), 0)
    Q_table[(tuple(tuple(row) for row in state), action)] += alpha*((r + gamma*Q_max) - Q_table[(tuple(tuple(row) for row in state), action)])

    state = next_state
    if(r == 1000):
      break
```
